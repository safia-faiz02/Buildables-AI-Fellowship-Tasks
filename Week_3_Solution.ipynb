### üìù Questions for Reflection

1. How is prompt chaining similar to solving a jigsaw puzzle?

Answer: It's similar to a jigsaw puzzle as each prompt is a little chunk or piece of solving the complex problem.

2. What‚Äôs the main difference between **iteration** and **chaining**?

Answer: Iteration is refining a one big single prompt, whereas chaining is all the prompts are linked to each other, completing one big picture together.

3. Why might prompt chaining be more effective than asking AI one big question?

Answer: as it breaks the task into manageable small chunks.

4. How can the **context window** of an AI model affect your chaining strategy?

Answer: context window is basically the memory model of an AI, how much past info can the AI remember to answer your question. if the context window is quite small (e.g: it remembers only the last five msgs), then you'd have to mould your chaining prompts accordingly.
